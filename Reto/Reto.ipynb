```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Cargar los datos desde el archivo CSV
data_train = pd.read_csv('train_procesados.csv')
data_test = pd.read_csv('tests_procesados.csv')

# Separar las características (X) y las etiquetas (y)
X_train = data_train.iloc[:, 1:]  # Todas las columnas excepto la primera son características
y_train = data_train.iloc[:, 0]   # La primera columna es la categoría

X_test = data_test.iloc[:, 1:]  # Todas las columnas excepto la primera son características
y_test = data_test.iloc[:, 0]   # La primera columna es la categoría
```

```python
# Contar la cantidad de muestras para cada clase
train_class_counts = data_train.iloc[:, 0].value_counts()
test_class_counts = data_test.iloc[:, 0].value_counts()

# Desplegar la cantidad de muestras por clase
print("Cantidad de muestras de entrenamiento:")
train_class_counts = train_class_counts.sort_index()
print(train_class_counts)

print("\nCantidad de muestras de test:")
test_class_counts = test_class_counts.sort_index()
print(test_class_counts)
```

```python
# Crear un modelo SVC
model = SVC(class_weight='balanced')

# Entrenar el modelo
model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Calcular la tabla de confusión normalizada
conf_matrix = confusion_matrix(y_test, y_pred, normalize='true')

# Desplegar la tabla de confusión normalizada usando seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt=".2f", cmap="Blues", cbar=False)
plt.title("Normalized Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

## Red Neuronal

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Cargar los datos
data_train = pd.read_csv('train_orig_proc.csv')
data_test = pd.read_csv('test_orig_proc.csv')

# Separar las características (X) y las etiquetas (y)
X_train = data_train.iloc[:, 1:].values
y_train = data_train.iloc[:, 0].values
X_test = data_test.iloc[:, 1:].values
y_test = data_test.iloc[:, 0].values

# Escalar los datos
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convertir las etiquetas a formato categórico (One-Hot Encoding)
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# Definir el modelo de red neuronal
model = Sequential()
model.add(Dense(X_train.shape[1]*3, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(X_train.shape[1]*3, activation='relu'))
model.add(Dense(y_train_cat.shape[1], activation='softmax'))  # Softmax para clasificación multiclase

# Compilar el modelo
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entrenar el modelo
model.fit(X_train, y_train_cat, epochs=2000, batch_size=2**16, validation_data=(X_test, y_test_cat))

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test_cat)
print(f'Loss: {loss}, Accuracy: {accuracy}')

# Predecir las etiquetas para el conjunto de prueba
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = y_test

# Calcular la matriz de confusión normalizada
cm = confusion_matrix(y_true_classes, y_pred_classes, normalize='true')

# Mostrar la matriz de confusión normalizada
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusión Normalizada')
plt.show()
```

## Busqueda de hipeparametros SVC (Tardo demasiado)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Cargar los datos desde el archivo CSV
data_train = pd.read_csv('train_orig_proc.csv')
data_test = pd.read_csv('test_orig_proc.csv')

# Separar las características (X) y las etiquetas (y)
X_train = data_train.iloc[:, 1:]  # Todas las columnas excepto la primera son características
y_train = data_train.iloc[:, 0]   # La primera columna es la categoría

X_test = data_test.iloc[:, 1:]  # Todas las columnas excepto la primera son características
y_test = data_test.iloc[:, 0]   # La primera columna es la categoría
```

```python


# Definir el rango de hiperparámetros a probar
param_grid = {
    'C': [0.1, 1, 10, 100],  # Parámetro de regularización
    'kernel': ['linear', 'rbf', 'poly'],  # Tipo de núcleo
    'gamma': ['scale', 'auto'],  # Parámetro del núcleo RBF
}

# Crear el modelo SVC
svc = SVC(class_weight='balanced')

# Configurar GridSearchCV
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)

# Entrenar el modelo con GridSearchCV
grid_search.fit(X_train, y_train)

# Obtener el mejor modelo
best_model = grid_search.best_estimator_

# Hacer predicciones en el conjunto de prueba
y_pred = best_model.predict(X_test)

# Evaluar el modelo
print("Best Parameters:", grid_search.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Calcular la tabla de confusión normalizada
conf_matrix = confusion_matrix(y_test, y_pred, normalize='true')

# Desplegar la tabla de confusión normalizada usando seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt=".2f", cmap="Blues", cbar=False)
plt.title("Normalized Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
import numpy as np
import pandas as pd
import joblib

# Cargar el modelo entrenado desde el archivo
modelo = joblib.load('Clasificador.pkl')

# Cargar la nueva muestra desde un archivo CSV (sin la columna de categoría)
nueva_muestra = pd.read_csv('Muestras/Muestra_cat4.csv', header=None)

# Definir los parámetros del preprocesamiento
num_frecuencias = 100
umbral = 0.04
n_partes = 20
frecuencia_muestreo = 125  # Hz
intervalo_tiempo = 1 / frecuencia_muestreo

# Inicializar listas para almacenar los resultados del preprocesamiento
amplitudes = []
desfases = []
contadores_derivada = []
muestras_originales = []

# Iterar sobre cada fila (muestra) en el archivo de la nueva muestra
for i in range(len(nueva_muestra)):
    fila = nueva_muestra.iloc[i, :].values  # Selecciona todas las columnas como muestra original

    # Guardar la muestra original
    muestras_originales.append(fila)

    # Realiza la Transformada de Fourier
    transformada = np.fft.fft(fila, num_frecuencias)

    # Calcula la amplitud y el desfase
    amplitud = np.abs(transformada)
    desfase = np.angle(transformada)

    # Añadir los resultados a las listas de amplitud y desfase
    amplitudes.append(amplitud)
    desfases.append(desfase)

    # Calcular la segunda derivada
    primera_derivada = np.diff(fila)
    segunda_derivada = np.diff(primera_derivada)

    # Dividir la señal en partes iguales
    longitud_segmento = len(segunda_derivada) // n_partes
    segunda_derivada = segunda_derivada[:longitud_segmento * n_partes]  # Ajustar el tamaño

    # Contar cuántas veces la segunda derivada supera el umbral en cada parte
    contadores = []
    for j in range(n_partes):
        segmento = segunda_derivada[j * longitud_segmento:(j + 1) * longitud_segmento]
        contador_segmento = np.sum(np.abs(segmento) > umbral)
        contadores.append(contador_segmento)

    # Añadir los resultados a la lista de contadores
    contadores_derivada.append(contadores)

# Crear DataFrames para las amplitudes, desfases y contadores de la segunda derivada
amplitudes_df = pd.DataFrame(amplitudes, columns=[f'Amplitud_{i+1}' for i in range(num_frecuencias)])
desfases_df = pd.DataFrame(desfases, columns=[f'Desfase_{i+1}' for i in range(num_frecuencias)])
contadores_df = pd.DataFrame(contadores_derivada, columns=[f'Contador_Segunda_Derivada_{i+1}' for i in range(n_partes)])

# Crear la columna de tiempos para la muestra original
n_muestras = len(nueva_muestra.columns)  # Número de muestras en cada fila (excluyendo la última columna)
tiempos = [f'T_{i * intervalo_tiempo:.3f}' for i in range(n_muestras)]

# Crear DataFrame para las muestras originales
muestras_df = pd.DataFrame(muestras_originales, columns=tiempos)

# Combinar los resultados del preprocesamiento en un solo DataFrame
nueva_muestra_procesada = pd.concat([amplitudes_df, desfases_df, contadores_df, muestras_df], axis=1)

# Obtener las probabilidades de cada clase
probabilidades = modelo.predict_proba(nueva_muestra_procesada)

# Mostrar la probabilidad para cada clase
print("Probabilidades de cada clase:", probabilidades[0])

# Mostrar la clase con la mayor probabilidad
print("La clase predicha para la nueva muestra es:", modelo.classes_[probabilidades[0].argmax()])

```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Cargar los datos desde el archivo CSV
data_train = pd.read_csv('mitbih_train.csv', header=None)
data_test = pd.read_csv('mitbih_test.csv', header=None)

# Separar las características (X) y las etiquetas (y)
X_train = data_train.iloc[:, :-1]  # Todas las columnas excepto la ultima son características
y_train = data_train.iloc[:, -1]   # La ultima columna es la categoría

X_test = data_test.iloc[:, :-1]  # Todas las columnas excepto la ultima son características
y_test = data_test.iloc[:, -1]   # La ultima columna es la categoría

# Crear un modelo SVC
model = SVC(class_weight='balanced', probability=True)

# Entrenar el modelo
model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test)
```
